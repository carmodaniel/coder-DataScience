```python
import pandas as pd
```

# 1 - Leitura Base [Brazilian E-Commerce Public Dataset by Olist](https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce/data?select=olist_order_payments_dataset.csv)

```python
geoloc = pd.read_csv('/content/dataset/olist_geolocation_dataset.csv')
geoloc
```

```python
items = pd.read_csv('/content/dataset/olist_order_items_dataset.csv')
items
```

```python
payments = pd.read_csv('/content/dataset/olist_order_payments_dataset.csv')
payments
```

```python
reviews = pd.read_csv('/content/dataset/olist_order_reviews_dataset.csv')
reviews
```

```python
orders = pd.read_csv('/content/dataset/olist_orders_dataset.csv')
orders
```

```python
products = pd.read_csv('/content/dataset/olist_products_dataset.csv')
products
```

```python
sellers = pd.read_csv('/content/dataset/olist_sellers_dataset.csv')
sellers
```

```python
cat_tr = pd.read_csv('/content/dataset/product_category_name_translation.csv')
cat_tr
```

```python
customers = pd.read_csv('/content/dataset/olist_customers_dataset.csv')
customers
```

# 2 - Agregando pagamentos por pedido (**para não multiplicar linhas**)

```python
pay_agg = (payments
           .groupby('order_id', as_index=False)
           .agg(payment_value_total=('payment_value','sum'),
                payment_installments_max=('payment_installments','max'),
                payment_types=('payment_type', lambda s: ','.join(sorted(s.astype(str).unique()))),
                payment_cnt=('payment_type','count')))
```

#3 - Agregando reviews por pedido

```python
rev_agg = (reviews
           .groupby('order_id', as_index=False)
           .agg(review_score_mean=('review_score','mean'),
                review_score_cnt=('review_id','count'),
                review_latest=('review_creation_date','max')))
```

# 4 -  União de cat_tr ao products

```python
products = products.merge(cat_tr, on='product_category_name', how='left')
```

# 5 - Preparação de dimensões de cliente e vendedor
## Padronizar as colunas de clientes e vendedores (cidade, estado, zip_code_prefix)

```python
cust_dim = customers.rename(columns={'customer_zip_code_prefix':'zip_code_prefix',
                                     'customer_city':'city','customer_state':'state'})
sell_dim = sellers.rename(columns={'seller_zip_code_prefix':'zip_code_prefix',
                                   'seller_city':'city','seller_state':'state'})
```

# 6 - Agregando geolocalização por prefixo

```python
geo_agg = (geoloc
           .groupby('geolocation_zip_code_prefix', as_index=False)
           .agg(geo_lat=('geolocation_lat','mean'),
                geo_lng=('geolocation_lng','mean')))
```

# 7 - Inclusão  de latitude longitude para Cliente e Fornecedor


```python
cust_dim = cust_dim.merge(geo_agg, left_on='zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left', suffixes=('','_cust'))
sell_dim = sell_dim.merge(geo_agg, left_on='zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left', suffixes=('','_sell'))
```

# 8 - Montar fato analítico no nível do item

```python
df = (items
      .merge(orders, on='order_id', how='left')  # status e tempos do pedido
      .merge(products, on='product_id', how='left')  # atributos do produto + categoria traduzida
      .merge(sell_dim.add_prefix('seller_'), left_on='seller_id', right_on='seller_seller_id', how='left')  # vendedor
      .merge(customers, on='customer_id', how='left')  # chave cliente_id
      .merge(cust_dim.add_prefix('customer_'), left_on='customer_id', right_on='customer_customer_id', how='left') # localização cliente
      .merge(pay_agg, on='order_id', how='left')  # pagamentos agregados
      .merge(rev_agg, on='order_id', how='left'))  # reviews agregadas

# Exibição da saída
df
```

# 9 - Selecção das colunas-chave para análise

```python
cols_keep = [
    # chaves
    'order_id','order_item_id','product_id','seller_id','customer_id',
    # métricas item
    'price','freight_value',
    # tempos do pedido
    'order_status','order_purchase_timestamp','order_approved_at',
    'order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date',
    # produto
    'product_category_name','product_category_name_english','product_weight_g','product_length_cm',
    'product_height_cm','product_width_cm',
    # cliente
    'customer_city_x','customer_state_y',
    'customer_geo_lat','customer_geo_lng',
    # vendedor
    'seller_city','seller_state',
    'seller_geo_lat','seller_geo_lng',
    # pagamentos agregados
    'payment_value_total','payment_installments_max','payment_types','payment_cnt',
    # reviews agregadas
    'review_score_mean','review_score_cnt','review_latest'
]
df_final = df[cols_keep]
df_final.to_csv('/content/dataset/df_final-olist_analitico_items.csv', index=False)
```

# Graficos

```python
# Top 10 categorias por receita
## Esse gráfico ajuda a identificar as categorias com maior contribuição de receita e direcionar análises de margem e estoque
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt

df_final['receita_item'] = df_final['price'] + df_final['freight_value']
top_cat = (df_final
           .groupby('product_category_name_english', as_index=False)['receita_item']
           .sum()
           .nlargest(10, 'receita_item'))

plt.figure(figsize=(10,5))
sns.barplot(data=top_cat, x='receita_item', y='product_category_name_english', palette='Blues_r')
plt.title('Top 10 categorias por receita')
plt.xlabel('Receita total (preço + frete)')
plt.ylabel('Categoria')
plt.tight_layout()
plt.show()
```

```python
#  Lead time de entrega por UF do cliente
## Boxplot do tempo entre compra e entrega por estado do cliente, evidenciando medianas e caudas (atrasos) regionais
df_final['order_purchase_timestamp'] = pd.to_datetime(df_final['order_purchase_timestamp'])
df_final['order_delivered_customer_date'] = pd.to_datetime(df_final['order_delivered_customer_date'])

df_ok = df_final.dropna(subset=['order_purchase_timestamp','order_delivered_customer_date'])
df_ok['lead_time_dias'] = (df_ok['order_delivered_customer_date'] - df_ok['order_purchase_timestamp']).dt.days

top_states = (df_ok['customer_state_y'].value_counts().head(12).index)  # focar UFs mais frequentes
df_plot = df_ok[df_ok['customer_state_y'].isin(top_states)]

plt.figure(figsize=(12,5))
sns.boxplot(data=df_plot, x='customer_state_y', y='lead_time_dias')
plt.title('Lead time (dias) por UF do cliente')
plt.xlabel('UF')
plt.ylabel('Dias entre compra e entrega')
plt.tight_layout()
plt.show()
```

```python
# Atraso vs. satisfação (review_score)
## Relação entre atraso (entregar após a estimativa) e a média de review_score, validando impacto do prazo na satisfação.
df_final['order_estimated_delivery_date'] = pd.to_datetime(df_final['order_estimated_delivery_date'])

df_rev = df_final.dropna(subset=['order_delivered_customer_date','order_estimated_delivery_date','review_score_mean'])
df_rev['late_flag'] = (df_rev['order_delivered_customer_date'] > df_rev['order_estimated_delivery_date']).astype(int)

avg_score = df_rev.groupby('late_flag')['review_score_mean'].mean().reset_index()
avg_score['late_flag'] = avg_score['late_flag'].map({0:'No prazo', 1:'Atrasado'})

plt.figure(figsize=(6,4))
sns.barplot(data=avg_score, x='late_flag', y='review_score_mean', palette='Set2')
plt.title('Satisfação média por status de atraso')
plt.xlabel('Status de entrega')
plt.ylabel('Review score médio')
plt.ylim(0,5)
plt.tight_layout()
plt.show()
```

```python
# Frete relativo por categoris
## Mostra o peso do frete no valor total por categoria, útil para política de frete grátis e precificação
df_final['valor_total_item'] = df_final['price'] + df_final['freight_value']
df_fre = df_final[df_final['valor_total_item'] > 0].copy()
df_fre['frete_rel'] = df_fre['freight_value'] / df_fre['valor_total_item']

top_cat2 = (df_fre.groupby('product_category_name_english')['valor_total_item'].sum()
                 .nlargest(10).index)
df_plot2 = df_fre[df_fre['product_category_name_english'].isin(top_cat2)]

plt.figure(figsize=(10,5))
sns.boxplot(data=df_plot2, x='product_category_name_english', y='frete_rel')
plt.title('Proporção do frete no total por categoria (Top 10 por receita)')
plt.xlabel('Categoria')
plt.ylabel('Frete / Total do item')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#  Receita ao longo do tempo (mensal)
## Série temporal da receita para avaliar sazonalidade, crescimento e choques operacionais
df_ts = df_final.dropna(subset=['order_purchase_timestamp']).copy()
df_ts['mes'] = df_ts['order_purchase_timestamp'].dt.to_period('M').dt.to_timestamp()
receita_mensal = df_ts.groupby('mes')['receita_item'].sum().reset_index()

plt.figure(figsize=(10,4))
sns.lineplot(data=receita_mensal, x='mes', y='receita_item', marker='o')
plt.title('Receita mensal (preço + frete)')
plt.xlabel('Mês')
plt.ylabel('Receita')
plt.tight_layout()
plt.show()
```
